# Word Clouds

## Getting Started

```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

## Load Data

```python
clean_df = pd.read_csv('integrated_clean.csv', index_col=0, header=0)
annotated_df = clean_df[clean_df['manual_annotation'] == True].copy()
print(annotated_df.shape)
annotated_df.drop(columns=['sent_id', 'title', 'focus_sentence', 'gave_delta',
       'received_delta', 'manual_annotation', 'comment_length',
       'clean_sentence', 'word_count'], inplace=True)
annotated_df.head()
```

```python
clean_df.keys()
```

```python
melted_aspect = annotated_df.melt(['LDA_ready'], 
                                    var_name='aspect', # variable変数の変数（カラム）名
                                    value_name='opinion') # value変数の変数（カラム）名
print("before", melted_aspect.shape)
melted_aspect = melted_aspect[melted_aspect["opinion"] != 'None']
print("after", melted_aspect.shape)
melted_aspect.head()
```

## Visualize

```python
from PIL import Image

stopwords = set(STOPWORDS)
def create_wordcloud(font_path, extra_stopwords, tag, save_flag, colormap):
    stopwords.update(extra_stopwords)
    if tag in ['humanity', 'science', 'economy', 'politics']:
        text = ' '.join(
            s for s in melted_aspect[melted_aspect['aspect']==f'aspect_{tag}']['LDA_ready'].tolist())
        img_color = np.array(Image.open('cloud.png'))
    elif tag in ['concerned', 'neutral', 'conflicted', 'unconcerned']:
        text = ' '.join(
            s for s in melted_aspect[melted_aspect['opinion']==tag]['LDA_ready'].tolist())
        img_color = np.array(Image.open('oval.png'))
    wordcloud = WordCloud(font_path=font_path, 
                          stopwords=stopwords, 
                          random_state=523,
                          colormap=colormap,
                          mask = img_color,
                          background_color='white',
                          mode = "RGBA",
                         ).generate(text)
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
    if save_flag:
        wordcloud.to_file(f"./{tag}.png")
```

```python
font_path = '/System/Library/Fonts/NewYork.ttf'
extra_stopwords = ["make", "much", "thing", "go", "will", "issue", "going", "think", "using", "climate change", "many", "problem", "climate", "change"]
```

```python
tag = 'humanity'
create_wordcloud(font_path, extra_stopwords, tag, False, 'magma')
```
